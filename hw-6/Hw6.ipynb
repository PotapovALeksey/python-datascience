```python
def h(x, w):
  return x @ w


def gradient_step(X, y, w, learning_rate):
  m = X.shape[1]
  error = h(X, w) - y

  gradient = (X.T @ error) / m

  return w - learning_rate * gradient

def gradient(X, y):
  m, n = X.shape
  w = np.zeros(n)
  learning_rate = 0.0001
  epochs = 100000



  for epoch in range(epochs):
      w = gradient_step(X, y, w, learning_rate)

  return w


features = ['area', 'bedrooms', 'bathrooms', 'price']
scaled_data = StandardScaler().fit_transform(df[features])
scaled_df = pd.DataFrame(scaled_data, columns=features)

X = scaled_df[['area', 'bedrooms', 'bathrooms']]
y = scaled_df['price']
X_b = np.c_[np.ones((X.shape[0],1)), X]

gradient_w = gradient(X_b, y)

gradient_w
```

```python
A = X_b.T @ X_b
b = X_b.T @ y
analytics_w = np.linalg.inv(A) @ b

analytics_w
```

```python
model = LinearRegression().fit(X, y)

np.insert(model.coef_, 0, model.intercept_)
```
